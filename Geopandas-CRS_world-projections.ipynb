```python
import geopandas as gpd 

# Import CRS class from pyproj library (Python wrapper for conducting coordinate transformations)
from pyproj import CRS

# Comparing the types of projections plotting the data
import matplotlib.pyplot as plt

```

```python
# Initialize the CRS class for epsg code 3035:
crs_object = CRS.from_epsg(3035)
crs_object

# # Retrive CRS information in WKT format
crs_wkt = crs_object.to_wkt()
# print(crs_wkt)


# Retrieve EPSG code from WKT text
epsg = CRS(crs_wkt).to_epsg()
print(epsg)
```

```python
# GLOBAL MAP PROJECTIONS
    # I am going to try different map projections

fp2 = "L2_data/ne_110m_admin_0_countries/ne_110m_admin_0_countries.shp"

admin = gpd.read_file(fp2)

admin.crs
```

```python
#Setting fig size 
plt.rcParams['figure.figsize'] = [14,7]

#Plot in original crs
admin.plot()
plt.title("WGS84")
```

```python
# Projection in web Mercator (equal area), 3785
web_mercator = CRS.from_epsg(3785)

#Re-projecting and ploting
admin.to_crs(web_mercator).plot()

# Removing x and y axis
plt.axis('off')
plt.title(" Web Mercator")
```
